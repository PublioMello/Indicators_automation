```python
import pathlib
import pandas as pd
import win32com.client as win32
```

```python

# Get the sheet with the sales 
sales_sheet = pd.read_excel(r'Bases de Dados/Vendas.xlsx')

# Get the stores and teh emails
store_sheet = pd.read_csv(r'Bases de Dados/Lojas.csv',sep=';', encoding='latin-1')
email_sheet = pd.read_excel(r'Bases de Dados/Emails.xlsx')

# Include the name og the store in Sales
sales_sheet = sales_sheet.merge(store_sheet, on='ID Loja')

# add the email to the store sheet
store_sheet = store_sheet.merge(email_sheet, on='Loja')

display(store_sheet)

display(sales_sheet)
```

```python
#used this to replace the original email to mine to see if its working
new_store=store_sheet
store_sheet["E-mail"] = store_sheet['E-mail'].replace(
    to_replace=r'pythonimpressionador\+([a-zA-Z0-9._%+-])',
    value=r'publio.lima.mello+\1',
    regex=True)
#display(new_store)
display(store_sheet)

#used this to replace the original email to mine to see if its working

email_sheet["E-mail"] = email_sheet['E-mail'].replace(
    to_replace=r'pythonimpressionador\+([a-zA-Z0-9._%+-])',
    value=r'publio.lima.mello+\1',
    regex=True)
#display(new_store)
display(email_sheet)
```

```python
# Creating tables for each store
dic_stores = {} 
for store in store_sheet['Loja']:
    dic_stores[store] = sales_sheet.loc[sales_sheet['Loja']==store, :]

# Test to see if its working
display(dic_stores['Bourbon Shopping SP'])
```

```python
# Get the last day
last_day = sales_sheet['Data'].max()
print(last_day)
print('{}/{}'.format(last_day.day,last_day.month))
```

```python
# Identify if the paste already exists

path_backup = pathlib.Path(r'Backup Arquivos Lojas')

# get all the objects that are inside of the directory
files_folder_backup = path_backup.iterdir()

#create a list with all the files that exists in the folder
list_folder_names_backup =[file.name for file in files_folder_backup] 

print(list_folder_names_backup)

for store in dic_stores: #get the name of the store inside of the dictionary
    if store not in list_folder_names_backup:# See if the name of the store of the dictonary is in the list of folders
        new_folder = path_backup / store # If not create a new folder
        new_folder.mkdir()
        print("The folder was created : {}".format(store))

    # Creating the name of the file
    file_name= '{}_{}_{}.xlsx'.format(last_day.day,last_day.month,store)
    print(file_name)
    #creating the path where its gonna be saved
    file_path = path_backup / store / file_name
    # Create a excel file with the dictonary and save in the right paste
    dic_stores[store].to_excel(file_path)
   
```

```python
# Setting goals
daily_revenue_target = 1000
year_revenue_target = 1650000
target_qauntityproduct_day = 4
target_qauntityproduct_year = 120
target_aov_day = 500
target_aov_year = 500
```

### Website used to create the body of the email in HTML

https://www.w3schools.com/html/tryit.asp?filename=tryhtml_table_intro

```python
i=0
#for store in dic_stores:
for store in dic_stores:
  sales_store = dic_stores[store] # get the sales of the store
  sales_day = sales_store.loc[sales_store['Data']==last_day,:] # get the sales of the store per day
  revenue_year = sales_store['Valor Final'].sum() #sum the sales of the store
  revenue_day = sales_day['Valor Final'].sum()    #sum the sales of the store per day

  # Diversity of products year
  amount_product_year = sales_store['Produto'].unique()

  # Diversity of products day
  amount_product_day = sales_day['Produto'].unique() 

  # Average order value (AOV) year
  sales_value = sales_store.groupby('Código Venda').sum()
  aov_year = sales_value['Valor Final'].mean()
  # Average order value (AOV) day
  sales_value_day = sales_day.groupby('Código Venda').sum()
  aov_day = sales_value_day['Valor Final'].mean()

  # Starting the email
  outlook = win32.Dispatch('outlook.application')
  mail = outlook.CreateItem(0)
  name = store_sheet.loc[store_sheet['Loja']==store, 'Gerente'].values[0] # Get the name of the manager from the storesheet
  mail.To = store_sheet.loc[store_sheet['Loja']==store, 'E-mail'].values[0] # Get the Email from the storesheet
  mail.Subject = f'OnePage Dia {last_day.day}/{last_day.month} - Loja {store}' #create the subject with the date and store 
  # Conditions for the colors
  if revenue_day >= daily_revenue_target:
      cor_fat_dia = 'green'
  else:
      cor_fat_dia = 'red'
  if revenue_year >= year_revenue_target:
      cor_fat_ano = 'green'
  else:
      cor_fat_ano = 'red'
  if len(amount_product_day) >= target_qauntityproduct_day:
      cor_qtde_dia = 'green'
  else:
      cor_qtde_dia = 'red'
  if len(amount_product_year) >= target_qauntityproduct_year:
      cor_qtde_ano = 'green'
  else:
      cor_qtde_ano = 'red'
  if aov_day >= target_aov_day:
      cor_ticket_dia = 'green'
  else:
      cor_ticket_dia = 'red'
  if aov_year >= target_aov_year:
      cor_ticket_ano = 'green'
  else:
      cor_ticket_ano = 'red'

#Email in HTMl used the website previously showed
  mail.HTMLBody = f'''
  <p>Good morning, {name}</p>

<p>Yesterday's results <strong>({last_day.day}/{last_day.month})</strong> for <strong>Store {store}</strong> were:</p>

<table>
  <tr>
    <th>Indicator</th>
    <th>Day Value</th>
    <th>Day Target</th>
    <th>Day Scenario</th>
  </tr>
  <tr>
    <td>Revenue</td>
    <td style="text-align: center">R${revenue_day:.2f}</td>
    <td style="text-align: center">R${daily_revenue_target:.2f}</td>
    <td style="text-align: center"><font color="{cor_fat_dia}">◙</font></td>
  </tr>
  <tr>
    <td>Product Diversity</td>
    <td style="text-align: center">{len(amount_product_day)}</td>
    <td style="text-align: center">{target_qauntityproduct_day}</td>
    <td style="text-align: center"><font color="{cor_qtde_dia}">◙</font></td>
  </tr>
  <tr>
    <td>Average Order Value</td>
    <td style="text-align: center">R${aov_day:.2f}</td>
    <td style="text-align: center">R${target_aov_day:.2f}</td>
    <td style="text-align: center"><font color="{cor_ticket_dia}">◙</font></td>
  </tr>
</table>
<br>
<table>
  <tr>
    <th>Indicator</th>
    <th>Year Value</th>
    <th>Year Target</th>
    <th>Year Scenario</th>
  </tr>
  <tr>
    <td>Revenue</td>
    <td style="text-align: center">R${revenue_year:.2f}</td>
    <td style="text-align: center">R${year_revenue_target:.2f}</td>
    <td style="text-align: center"><font color="{cor_fat_ano}">◙</font></td>
  </tr>
  <tr>
    <td>Product Diversity</td>
    <td style="text-align: center">{len(amount_product_year)}</td>
    <td style="text-align: center">{target_qauntityproduct_year}</td>
    <td style="text-align: center"><font color="{cor_qtde_ano}">◙</font></td>
  </tr>
  <tr>
    <td>Average Order Value</td>
    <td style="text-align: center">R${aov_year:.2f}</td>
    <td style="text-align: center">R${target_aov_year:.2f}</td>
    <td style="text-align: center"><font color="{cor_ticket_ano}">◙</font></td>
  </tr>
</table>

<p>Attached is the spreadsheet with all the data for more details.</p>

<p>If you have any questions, please feel free to contact me.</p>
<p>Best regards, Publio</p>
  '''

  # Attachments (as many as we want):
  attachment  = pathlib.Path.cwd() / path_backup / store / f'{last_day.day}_{last_day.month}_{store}.xlsx'
  mail.Attachments.Add(str(attachment))

  mail.Send()
  print('E-mail da Loja {} enviado'.format(store))
  i+=1

print(f"There were sent {i} E-mails")
```

```python
# Group by 'Loja' and sum 'Valor Final'
billing_store_year = sales_sheet.groupby('Loja')[['Loja','Valor Final']].sum()
# Sort by 'Valor Final' in descending order
billing_store_year = billing_store_year.sort_values(by='Valor Final',ascending=False)
# Format the 'Valor Final' column to include thousand separators
billing_store_year['Valor Final'] = billing_store_year['Valor Final'].apply(lambda x: f'{x:,.2f}')
display(billing_store_year)

file_billing_store = '{}_{}_Anual Ranking.xlsx'.format(last_day.day,last_day.month)
billing_store_year.to_excel(r'Backup Arquivos Lojas\{}'.format(file_billing_store))

# Filter the data for the last day
billing_store_day = sales_sheet.loc[sales_sheet['Data']==last_day,:]
# Group by 'Loja' and sum 'Valor Final'
billing_store_day = billing_store_day.groupby('Loja')[['Loja','Valor Final']].sum()
# Sort by 'Valor Final' in descending order
billing_store_day = billing_store_day.sort_values(by='Valor Final',ascending=False) 
# Format the 'Valor Final' column to include thousand separators
billing_store_day['Valor Final'] = billing_store_day['Valor Final'].apply(lambda x: f'{x:,.0f}'.replace(',', '.'))
display(billing_store_day)

file_billing_store_day = '{}_{}_Daily Ranking.xlsx'.format(last_day.day,last_day.month)
billing_store_day.to_excel(r'Backup Arquivos Lojas\{}'.format(file_billing_store_day))
```
